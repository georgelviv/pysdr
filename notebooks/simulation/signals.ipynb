# Signals

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
sample_rate = 1000
freq = 5
t = np.linspace(0.0, 2.0, sample_rate)
s1 = np.sin(np.pi * t * 5)
s2 = np.sin(np.pi * t * 10)
s3 = np.sin(np.pi * t * 15)
sum = s1 + s2 + s3

figsize = (5, 2)
fig, axs = plt.subplots(2, 2)
fig.suptitle('Sin waves')
axs[0, 0].plot(t, s1)
axs[0, 0].set_title('5 Hz')
axs[0, 1].plot(t, s2)
axs[0, 1].set_title('10 Hz')
axs[1, 0].plot(t, s3)
axs[1, 0].set_title('15 Hz')
axs[1, 1].plot(t, sum)
axs[1, 1].set_title('Sum')

for ax in axs.flat:
    ax.label_outer()

fig.show()
```

```python
from scipy.fft import fft, fftfreq

yf = fft(sum)
xf = fftfreq(s2.size, 1 / sample_rate)

plt.figure(figsize=figsize)
plt.plot(xf, np.abs(yf))
plt.grid(True)
plt.xlim(0, 20)
plt.show()
```
