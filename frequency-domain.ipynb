```python
import numpy as np
from matplotlib import pyplot as plt

Fs = 1
N = 100

t = np.arange(N)
s = np.sin(0.15*2*np.pi*t)
s = s * np.hamming(100)
S = np.fft.fftshift(np.fft.fft(s))
S_mag = np.abs(S)
S_phase = np.angle(S)
f = np.arange(Fs/-2, Fs/2, Fs/100)

plt.figure(0)
plt.plot(t, s, '.-')
plt.show()
plt.figure(1)
plt.plot(f, S_mag, '.-')
plt.figure(2)
plt.plot(f, S_phase, '.-')
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt


sample_rate = 1e6
t = np.arange(1024*1000)/sample_rate
f = 50e3
x = np.sin(2*np.pi*f*t) + 0.2 * np.random.randn(len(t))

fft_size = 1024
num_rows = len(x) // fft_size
spectrogram = np.zeros((num_rows, fft_size))

for i in range(num_rows):
  f = np.abs(np.fft.fftshift(np.fft.fft(x[i*fft_size:(i+1)*fft_size])))
  row = 10*np.log10(f**2)
  spectrogram[i,:] = row

plt.figure(0)
plt.plot(np.arange(200), x[:200], '-')

plt.figure(1)
plt.imshow(spectrogram, aspect='auto', extent=[sample_rate/-2/1e6, sample_rate/2/1e6, len(x)/sample_rate, 0])
plt.xlabel('Frequency [MHz]')
plt.ylabel('Time [s]')
plt.show()
print(t)
```
